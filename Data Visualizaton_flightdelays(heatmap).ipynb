```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
flight_filepath="C:\\Users\\ASUS\\Desktop\\Datasets\\flight_delays.csv"
flight_data=pd.read_csv(flight_filepath,index_col="Month")
flight_data.head()
```

```python
plt.figure(figsize=(16,6))
plt.title("Average arrival delay for each airline by Month")
sns.heatmap(data=flight_data,annot=True)
plt.xlabel("Airline")
```

```python
#the months toward the end of the year (especially months 9-11) appear relatively dark for all airlines. This suggests that airlines are better (on average) at keeping schedule during these months!
```
