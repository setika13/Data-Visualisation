```python
#importing libraries
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
#Specifying the file path
museum_filepath="C:\\Users\\ASUS\\desktop\\Datasets\\museum_visitors.csv"
museum_data=pd.read_csv(museum_filepath,index_col="Date",parse_dates=True)
```

```python
#Reviewing the last 5 details
museum_data.tail()
```

```python
#Creating line chart to show number of visitors evolved over time
plt.figure(figsize=(16,6))
sns.lineplot(data=museum_data)
plt.title("Showing number of visitors")

```
